5.2 Classication pour la détection de la **tromperie**

1. Partitionner les données selon l'approche K-folds

```python
from sklearn.model_selection import KFold
KF = KFold(n_splits=6,random_state=42,shuffle=True)
X=data.drop(columns='class',axis=1)
Y=class_tromperie
for train_index, test_index in KF.split(X):
  X_train, X_test = X.iloc[train_index], X.iloc[test_index]
  Y_train, Y_test = Y.iloc[train_index], Y.iloc[test_index]

```

```python
X_train = X_train.loc[:,~X_train.columns.duplicated()]
X_test = X_test.loc[:,~X_test.columns.duplicated()]

X_train
```

```python
X_test
```

```python
Y_train.head
```

2. Modèle d'apprentissage de la classification (nous avons combiné deux modèles ML) avec la technique de l'empilement, c'est-à-dire la décision finale par vote majoritaire des modèles sélectionnées.

```python
from xgboost import XGBClassifier
X_G_B_classifier = XGBClassifier()
```

```python
from sklearn.svm import SVC 
SVM_Classifier=SVC(gamma='auto',kernel="linear")
```

```python
from sklearn.ensemble import RandomForestClassifier as RF
RF_classifier= RF(n_estimators=200,max_depth=100)
```

```python
from sklearn.ensemble import StackingClassifier
from sklearn.linear_model import LogisticRegression 
estimators = [('RF', RF_classifier),
          ('XGB', X_G_B_classifier),
          ('SVM', SVM_Classifier)]

Model = StackingClassifier(estimators=estimators, final_estimator=LogisticRegression())
```

```python
Model
```

```python
Model.fit(X_train, Y_train).score(X_test, Y_test)
```

```python
y_pred2=Model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,y_pred2)
```

```python
Model.score(X_train, Y_train)
```

3. Matrice de confusion + Evaluation des metrics.

```python
from sklearn.metrics import confusion_matrix
Y_predicted=Model.predict(X_train)
```

```python
confusion_matrix(Y_train,Y_predicted)
```

```python
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
```

```python
def get_accuracy(trained_model,X, Y):
    predicted = trained_model.predict(X)
    accuracy = np.mean(predicted == Y)
    return accuracy
```

```python
def get_F1(trained_model,X,Y):
    predicted=trained_model.predict(X)
    f1=f1_score(Y,predicted)
    return f1
```

```python
def get_recall(trained_model,X,Y):
    predicted=trained_model.predict(X)
    f1=f1_score(Y,predicted)
    return f1
```

```python
print("La précision de notre modèle est:",get_accuracy(Model,X_test,Y_test))
print("Le F1_Score de notre modèle est:",get_F1(Model,X_test,Y_test))
print("Le recall score de notre modèle est:",get_recall(Model,X_test,Y_test))
```

5.3 Amélioration des performances en optimisant les hyper-paramétres de la
technique de classication

```python
## Hyperparameter tunning with RandomCV search , GridSearch 
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV

param_RF = {"max_depth": [2,8, None],
                "max_features": [0.2,0.6,1.0],
                "n_estimators": [20, 60, 100,120] ,
                "max_samples": [0.5, 0.75, 1.0],
                # "bootstrap": [True, False]
            }
RF_Tunning=GridSearchCV(estimator=RF_classifier, param_grid=param_RF, scoring='accuracy', 
                       cv=5, refit=True, return_train_score=True)

param_SVM = {
        # Parameters that we are going to tune.
        'C': [0.1, 1, 10, 100, 1000], 
              'gamma': ["auto","scale"],
              'kernel': ['rbf',"linear"]} 

SVM_Tunning=GridSearchCV(estimator=SVM_Classifier, param_grid=param_SVM, scoring='accuracy', 
                       cv=6, refit=True, return_train_score=True) 


XGB_params={
    # Parameters that we are going to tune.
    'max_depth':[3,4,5,6,8,10,12,15],
    'n_estimators':[100,150],
    'gamma':[0.0,.1,0.2,0.3,0.4],
    'min_child_weight':[1,3,5,7],
    'learning_rate': [0.05,0.1,0.15,0.2,0.25,0.3],
    'colsample_bytree': [0.3,0.4,0.5,0.7]}

XGB_Tunning = RandomizedSearchCV(estimator=X_G_B_classifier, param_distributions=XGB_params,
                              n_iter=10, scoring='accuracy', cv=5,refit=True, n_jobs=-1)


```

```python
RF_Tunning.fit(X_train,Y_train)
```

```python
RF_Tunning.score(X_test,Y_test)
```

```python
RF_Tunning.best_params_
```

```python
SVM_Tunning.fit(X_train,Y_train).score(X_test,Y_test)
```

```python
SVM_Tunning.best_params_
```

```python
XGB_Tunning.fit(X_train,Y_train).score(X_test,Y_test)
```

```python
XGB_Tunning.best_params_
```

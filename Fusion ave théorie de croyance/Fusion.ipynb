1. Ré-écrire la fonction fournie en Python

```python
def belief(alpha_1,alpha_2,score_1,score_2):
  bet_1_yes=score_1*alpha_1+(1-alpha_1)/2
  bet_1_no=(1-score_1)*alpha_1+(1-alpha_1)/2
  bet_2_yes=score_2*alpha_2+(1-alpha_2)/2
  bet_2_no=(1-score_2)*alpha_2+(1-alpha_2)/2    
  
  Prob_high=bet_1_yes*bet_2_yes
  Prob_low=bet_1_yes*bet_2_no+bet_1_no*bet_1_yes
  Prob_null=bet_1_no*bet_2_no
  L=[]
  L.append(Prob_high)
  L.append(Prob_low)
  L.append(Prob_null)
  return(L)


```

2. Déterminer à quelle classe tromperie ou vérité appartient chaque combinaison testée selon la théorie de croyance pour αt = 0.8 et αv = 0.95

```python
L2=["High Delusion","Low Delusion","Veracity"]
alpha_1,alpha_2=0.8,0.95
Seq1=belief(alpha_1,alpha_2,0.8,0.95)
Seq2=belief(alpha_1,alpha_2,0.5,0.95)
Seq3=belief(alpha_1,alpha_2,0.5,0.6)
Seq4=belief(alpha_1,alpha_2,0.2,0.1)
```

**Fusion avec 3 modalités :**

 les probabilités des 3 niveaux de tromperie sont calculées selon les formules suivantes :
 image.png


```python
def belief_3M(alpha_t,alpha_v,alpha_a,score_t,score_v,score_a):
  bet_t_yes=score_t*alpha_t+(1-alpha_t)/2
  bet_t_no=(1-score_t)*alpha_t+(1-alpha_t)/2
  bet_v_yes=score_v*alpha_2+(1-alpha_v)/2
  bet_v_no=(1-score_v)*alpha_2+(1-alpha_v)/2    
  bet_a_yes=score_a*alpha_a+(1-alpha_a)/2
  bet_a_no=(1-score_a)*alpha_a+(1-alpha_a)/2
  Prob_high=bet_t_yes*bet_v_yes*bet_a_yes
  Prob_low=bet_v_yes*bet_a_yes*bet_t_no + bet_v_yes*bet_a_no*bet_t_no + bet_v_no*bet_a_yes*bet_t_yes + bet_v_no*bet_a_no*bet_t_yes+ bet_v_yes*bet_a_no*bet_t_yes+bet_v_no*bet_a_yes*bet_t_yes
  Prob_null=bet_t_no*bet_v_no*bet_a_no
  L=[]
  L.append(Prob_high)
  L.append(Prob_low)
  L.append(Prob_null)
  return(L)  
```

On choisisera 7 observations cette fois en respectant la distribution des observations sélectionnées pour le calcul de probabilité à priori.

```python
alpha_t,alpha_v,alpha_a=0.86,0.82,0.83
test_t=df_texte.sample(n=7,random_state=42)
test_v=df_video.reindex(test_t.index)
test_a=(df_audio.reindex(test_t.index)).iloc[:,0:35]

## score texte :
score_t_all_observations=classif_text.predict_proba(test_t.drop(columns=["Class"],axis=1))
## score video
score_v_all_observations=classif_video.predict_proba(test_v)
##score audio
score_a_all_observations=classif_audio.predict_proba(test_a)
```
